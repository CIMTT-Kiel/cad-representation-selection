# Median-Prediction-Baseline

```python
# third party imports
import pandas as pd
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np


# standard imports
from pathlib import Path

# Setup
PATH_TO_TESTSET = Path("/home/coder/shared/datasets/fabwave/5_model_input/test.csv")
PATH_TO_TRAINSET = Path("/home/coder/shared/datasets/fabwave/5_model_input/train.csv")

targets = ["faces", "volume", "edges", "vertices"]
```

```python
trainset_features_df = pd.read_csv(PATH_TO_TRAINSET)
testset_features_df = pd.read_csv(PATH_TO_TESTSET)
```

```python
# Median des Trainingssets berechnen
median_values = trainset_features_df[targets].median()

# Auf Testset anwenden
results = []
for target in targets:
    y_true = testset_features_df[target]
    y_pred = np.full(len(y_true), median_values[target])
    
    results.append({
        "target": target,
        "median": median_values[target],
        "MAE": mean_absolute_error(y_true, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_true, y_pred)),
        "R2": r2_score(y_true, y_pred)
    })

median_results_df = pd.DataFrame(results).set_index("target")
median_results_df
```
